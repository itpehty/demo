```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType

spark = SparkSession.builder.getOrCreate()

filePath = 'C:/data/trans/*'
schema = StructType([
                    StructField("regno", StringType(), True),
                    StructField("orgname", StringType(), True),
                    StructField("bizname", StringType(), True),
                    StructField("bizregno", StringType(), True),
                    StructField("crp_prv_cd", StringType(), True),
                    StructField("ceoname", StringType(), True),
                    StructField("tel", StringType(), True),
                    StructField("email", StringType(), True),
                    StructField("regdate", StringType(), True),
                    StructField("address", StringType(), True),
                    StructField("address_road", StringType(), True),
                    StructField("status", StringType(), True),
                    StructField("orgtel", StringType(), True),
                    StructField("salestype", StringType(), True),
                    StructField("product", StringType(), True),
                    StructField("domain", StringType(), True),
                    StructField("hostlocation", StringType(), True)
                ]) 
df=spark.read.csv(filePath,header=True,schema=schema)
df.createOrReplaceTempView("onl_mkt_biz_raw")
df.show()
```

```python
sqlDF1 = spark.sql('select count(1) from onl_mkt_biz_raw')
sqlDF1.show()

sqlDF2 = spark.sql('select regno,orgname,bizname,replace(bizregno,"-","") as bizregno,crp_prv_cd,ceoname,replace(tel,"-","") as tel,email,replace(regdate,"-","") as regdate,address,address_road,status,replace(orgtel,"-","") as orgtel,salestype,product,domain,hostlocation from onl_mkt_biz_raw')
sqlDF2.createOrReplaceTempView("onl_mkt_biz_raw")
sqlDF2.show()
```

```python
sqlDF3 = spark.sql('select bizregno from onl_mkt_biz_raw group by bizregno having count(1) > 1')
sqlDF3.createOrReplaceTempView("onl_mkt_biz_dup")
sqlDF3.show()
```

```python
sqlDF4 = spark.sql('select a.* from onl_mkt_biz_raw a where bizregno not in (select bizregno from onl_mkt_biz_dup)')
sqlDF4.show()
```
